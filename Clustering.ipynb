# Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv('cleaned_mall_customers.csv')
```

```python
#Creating values for the elbow
X = df.loc[:,["Age", "Annual_Income", "Spending_Score"]]
inertia = []
k = range(1,20)
for i in k:
    means_k = KMeans(n_clusters=i, random_state=0)
    means_k.fit(X)
    inertia.append(means_k.inertia_)
```

#Plotting the elbow

```python
# import matplot, seaborn and plotly for good experience visuals
import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(k , inertia , 'bo-')
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

# The number of clusters is equal to 5. Therefore we are selecting 5 as the number of clusters to divide our data in.
# OR we can Approximately elbow is formed at "5"

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = df[['Age', 'Annual_Income', 'Spending_Score']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=5, random_state=42) # from elbow meathod
df['Cluster'] = kmeans.fit_predict(scaled_features)

plt.figure(figsize=(10, 6))
sns.scatterplot(df, x='Annual_Income', y='Spending_Score', hue='Cluster', palette='viridis')
plt.title('Customer Segments')
plt.show()
```
