```python
import pandas as pd

df = pd.read_csv('sample_data.csv')

df.head(8)

# Check for missing values
x = df.isnull().sum()
y = df.info()
# print("NULL data counting wrt column:\n",x, "Information of csv without null and datatype\n",y)
# print(f"NULL data counting wrt column:\n{x}\nInformation of csv without null and datatype:\n{y}")
```

```python
# Remove rows with any missing values

delete_rows = df.dropna()
print(delete_rows.head(8))
```

```python
# Fill missing values with a specific value

# filled_value = df.fillna(0)  "ISSE null wale "0" fill ho gaye"
filled_value = df.fillna({
    'Age' : df['Age'].mean(),
    'Salary' : "me nhi bataaunga"
})
# filled_value = df.fillna(df.median())
# filled_value = df.fillna(df.mode())
# filled_value = df.fillna(df.max())
# filled_value = df.fillna(df.min())
# filled_value = df.fillna()
print(filled_value.head(8))
```

```python
# Forward fill method to propagate the next values backward
# Backward fill method to propagate the previous values forward

x = df.fillna(method='ffill')
y = df.fillna(method='bfill')
print("Forward fill:\n\n", x,"\n\n","Backwardfill\n\n", y)
```

```python
# Add duplicate rows for demonstration
df = pd.concat([df, df.iloc[[0]], df.iloc[[1]]], ignore_index=True)
print('Before removing duplicates:\n\n', df)

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print('\n\nAfter removing duplicates:\n\n', df_no_duplicates)
```

```python
# Replace incorrect values in the 'Department' column
from tabulate import tabulate

x = df.replace({'Department' : {'HR' : "Human Recource", 'Engineering' : 'Berozgaar', 'Sales' : 'Sales Department'}})
# print(tabulate(x, headers='keys', tablefmt='pretty'))
x.head(8)
```

```python
# Convert all department names to lowercase for consistency
df['Department'] = df['Department'].str.lower()
print(tabulate(df, headers='keys', tablefmt='pretty'))
```

```python
# Apply Min-Max normalization using the formula
df_normalized = df.copy()
for col in ['Age', 'Salary']:
    df_normalized[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# Print original and normalized values
print("Original DataFrame:")
print(df)
print("\nNormalized DataFrame:")
print(df_normalized)
```
